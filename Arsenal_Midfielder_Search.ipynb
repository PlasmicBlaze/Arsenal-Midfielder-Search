Over the years, we have looked at several players to cover one or both of the 6 and 8 roles. To begin this conversation, I'd like to establish a few things when it comes to recruitment as a basic set of beliefs I hold:

- Recruiting a small number of higher quality players is better than recruiting several lower quality additions to address the same needs.

- A younger player is great, but at the stage of Arsenal's project, any recruited player needs to be ready to play in the first team. Hence, an extremely promising prospect would not *usually* be an ideal recruit, especially if they are not deemed able to play 2000 minutes a season. My ideal recruitment age range is 21 (with a track record of experience and fitness) to 27 years old (if a player is that good and does not have a questionable fitness history).

- Recruiting hybrid players without a nailed down position, but rather for certain qualities they offer, is more ideal at this stage. It allows for a more "amorphous" distribution of roles around the team, more unpredictability in rotations, and more combinations in first XIs we field.

- When in doubt, the most impressive base profile (physical and technical) should be recruited. I give physical profiles a bit more weight for Arsenal's recruitment ideals specifically, because technicality often becomes a baseline requirement and physical returns see huge dividends in my view, especially with regard to player fitness. However, I *might* not always be consistent with this view on a case by case basis, because it is difficult to quantify a physicality vs technicality discussion, and there is always room for differing views.

- Finally, borrowing a bit from the previous point, fitness is paramount. The greatest ability is availability, and over the course of a gruelling season with ever increasing amounts of games, we need our players to be available.

It is quite difficult to talk about our recruitment in context without establishing what qualities we **seem** to desire in additions as a club:

- A transition treat, someone with the raw speed to trouble teams that push their backline high. Martinelli is the only such player we have at the moment.

- A central duelist to add legs to the midfield; someone who can defend transitions. Only Rice covers this base for us, especially with Partey losing athleticism with age.

- A player with ball-dominance; someone who can rack up touches of the ball and make quickfire passes to progress the ball. Jorginho and Partey are our options for this role, and both are ageing (over 30 years old), and have difficulties playing a large number of minutes.


The *easiest* ways to bring these qualities in, in theory, would be a winger (covering 1, 2) and a midfielder covering (3, 4).
```python
# Load necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
import io
```

```python
# Upload the data
file_path = 'LeftMid.xlsx'
df = pd.read_excel(io.BytesIO(uploaded[file_path]), sheet_name='Worksheet', skiprows=1)
# Use the first row as column headers
df.columns = df.iloc[0]
df = df.drop(0)
df
```

```python
# Drop columns with NaN as their title
df = df.loc[:, ~df.columns.isna()]

# Remove 'PPRk' column
df = df.drop(columns=['PPRk'])

# Remove duplicate columns
df = df.loc[:, ~df.columns.duplicated()]

df
```

```python
# Convert all columns except 'Player' and 'Pos' to numeric
for col in df.columns:
    if col not in ['Player', 'Pos', 'Comp']:
        df[col] = pd.to_numeric(df[col], errors='coerce')
```

```python
df
```

```python
# Select only numeric columns for clustering
numeric_cols = df.select_dtypes(include=[np.number])
numeric_cols
```

```python
# Remove columns with all NaN values
numeric_cols = numeric_cols.dropna(axis=1, how='all')

# Check for missing values and handle them
numeric_cols = numeric_cols.apply(lambda x: x.fillna(x.mean()), axis=0)

# Verify that there are no NaN values in the numeric columns
print("Checking for NaN values in numeric columns:")
print(numeric_cols.isna().sum())
```

```python
numeric_cols
```

```python
# Normalize the data
scaler = StandardScaler()
numeric_cols_normalized = pd.DataFrame(scaler.fit_transform(numeric_cols), columns=numeric_cols.columns)

# Verify that there are no NaN values in the normalized data
print("Checking for NaN values in normalized numeric columns:")
print(numeric_cols_normalized.isna().sum())
```

```python
# Determine the optimal number of clusters using the Elbow method
wss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(numeric_cols_normalized)
    wss.append(kmeans.inertia_)

plt.plot(range(1, 11), wss, marker='o')
plt.title('Elbow Method for Optimal Number of Clusters')
plt.xlabel('Number of clusters')
plt.ylabel('WSS')
plt.show()

# Set number of clusters (adjust based on the Elbow method)
num_clusters = 5

# Run k-means clustering
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df['cluster'] = kmeans.fit_predict(numeric_cols_normalized)

# Perform PCA for visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(numeric_cols_normalized)
pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2'])
pca_df['Player'] = df['Player']
pca_df['cluster'] = df['cluster']

# Plot the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df, x='PC1', y='PC2', hue='cluster', palette='viridis', s=100, alpha=0.7)
for line in range(0, pca_df.shape[0]):
    plt.text(pca_df.PC1[line], pca_df.PC2[line], pca_df.Player[line],
             horizontalalignment='left', size='medium',
             color='black', weight='semibold')
plt.title('Player Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend()
plt.show()

```

```python
# Evaluate the quality of the clustering with the silhouette score
silhouette_avg = silhouette_score(numeric_cols_normalized, df['cluster'])
print(f'Average silhouette width: {silhouette_avg:.2f}')

# Print players in each cluster
print("Players in each cluster:")
print(df[['Player', 'cluster']].sort_values(by='cluster'))

```

```python
# Filter players from Cluster 1 (change the cluster number as needed)
cluster_1_players = df[df['cluster'] == 2]

# Perform PCA for visualization on Cluster 1 players
numeric_cols_cluster_1 = cluster_1_players.select_dtypes(include=[np.number])
numeric_cols_cluster_1 = numeric_cols_cluster_1.apply(lambda x: x.fillna(x.mean()), axis=0)

# Normalize data for only Cluster 1 players
numeric_cols_cluster_1_normalized = pd.DataFrame(scaler.fit_transform(numeric_cols_cluster_1), columns=numeric_cols_cluster_1.columns)

# Perform PCA
pca_result_cluster_1 = PCA(n_components=2).fit_transform(numeric_cols_cluster_1_normalized)

# Verify PCA result for NaNs in Cluster 1
print("Checking for NaN values in PCA result for Cluster 1:")
print(pd.DataFrame(pca_result_cluster_1).isna().sum())

pca_df_cluster_1 = pd.DataFrame(pca_result_cluster_1, columns=['PC1', 'PC2'])
pca_df_cluster_1['Player'] = cluster_1_players['Player'].values  # Ensure player names are correctly aligned

# Plot the PCA results for Cluster 1 players
plt.figure(figsize=(10, 6))
sns.scatterplot(data=pca_df_cluster_1, x='PC1', y='PC2', s=100, alpha=0.7)
for line in range(0, pca_df_cluster_1.shape[0]):
    plt.text(pca_df_cluster_1.PC1[line], pca_df_cluster_1.PC2[line], pca_df_cluster_1.Player[line], horizontalalignment='left', size='medium', color='black', weight='semibold')
plt.title('PCA of Players in Cluster 1')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
